```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
import io

# Open the PDF file in binary mode
pdf_file = open('/content/drive/MyDrive/Colab Notebooks/Code_Clinic/Canvas8-GravitasGrp/sample.pdf', 'rb')

# Create a PDF resource manager object and a PDF page interpreter object
resource_manager = PDFResourceManager()
output_stream = io.StringIO()
laparams = LAParams()

# Create a PDF device object and a PDF interpreter object
device = TextConverter(resource_manager, output_stream, laparams=laparams)
interpreter = PDFPageInterpreter(resource_manager, device)

# Process each page in the PDF file
for page in PDFPage.get_pages(pdf_file):
    interpreter.process_page(page)

# Get the text from the output stream
text = output_stream.getvalue()

# Close the PDF file, device, and output stream
pdf_file.close()
device.close()
output_stream.close()

# Print the extracted text
print(text)
```

```python
import re

# Remove unwanted characters and symbols
clean_text = re.sub(r'[\n\t]+', ' ', text)   # Replace newlines and tabs with a space
clean_text = re.sub(r'[^\a-zA-Z0-9%\w\s]', '', clean_text)  # Remove non-alphanumeric characters
clean_text = re.sub(r'\s+', ' ', clean_text)   # Replace multiple spaces with a single space
clean_text = clean_text.lower()   # Convert text to lowercase

# Print the cleaned text
clean_text
```

```python
from transformers import BertForQuestionAnswering
```

```python
model = BertForQuestionAnswering.from_pretrained("deepset/bert-base-cased-squad2")
```

```python
 from transformers import AutoTokenizer
```

```python
tokenzier = AutoTokenizer.from_pretrained("deepset/bert-base-cased-squad2")
```

```python
  question = ["How much % increase to Butterfly's web chat support service in 2020?","How much is the price of food rise in Australia?"]
  question[0]
```

```python
tokenzier.encode(question[0],truncation=True,padding=True)
```

```python
from transformers import pipeline
```

```python
nlp = pipeline("question-answering",model=model,tokenizer=tokenzier)
```
